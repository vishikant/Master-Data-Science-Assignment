```python
#Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.
Matplotlib is a popular Python library used for data visualization. It provides a wide range of tools for creating static, animated, and interactive plots in Python. Matplotlib is particularly useful for generating plots and charts that are similar in style to MATLAB, making it an essential tool for data analysis, scientific research, and machine learning.

The `Pyplot` module within Matplotlib is a collection of functions that make it easy to create various types of plots. 

**Five plots** that can be created using `Pyplot` are:
1. **Line Plot**
2. **Bar Plot**
3. **Histogram**
4. **Scatter Plot**
5. **Pie Chart**
```

```python
#Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated dataplot a scatter plot.

#A scatter plot is a type of data visualization that displays individual data points on a two-dimensional graph, typically with one variable 
#on the x-axis and another on the y-axis. 
#It's used to observe relationships, patterns, or correlations between the two variables, making it ideal for exploratory data analysis.

import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.title("plotting of x and y")
plt.xlabel("x values")
plt.ylabel("y values")
```

```python
#Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
import numpy as np
x = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
plt.figure(figsize=(10,8))
plt.subplot(2,2,1)
plt.plot(x,y1,label="line1")
```

```python
plt.subplot(2, 2, 2)  # 2 rows, 2 columns, position 2
plt.plot(x, y2, label='Line 2', color='orange')
plt.title("Plot 2")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Plot 3
plt.subplot(2, 2, 3)  # 2 rows, 2 columns, position 3
plt.plot(x, y3, label='Line 3', color='green')
plt.title("Plot 3")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Plot 4
plt.subplot(2, 2, 4)  # 2 rows, 2 columns, position 4
plt.plot(x, y4, label='Line 4', color='red')
plt.title("Plot 4")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()
```

```python
#Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

#A bar plot (or bar chart) is a type of data visualization used to represent categorical data with rectangular bars. 
#Each bar's height (or length) is proportional to the value it represents, allowing for easy comparison between different categories.

plt.bar(company,profit)
```

```python
#Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

A box plot (or box-and-whisker plot) is a standardized way of displaying the distribution of data based on a five-number summary: minimum, 
first quartile (Q1), median (Q2), third quartile (Q3), and maximum. The "box" shows the interquartile range (IQR) where the middle 50% of 
the data lies, while the "whiskers" extend to the minimum and maximum values, excluding outliers. Outliers, if any, are often plotted as 
individual points.


Why is it Used?
Box plots are used to visually summarize the distribution, central tendency, and variability of a dataset. They are particularly useful for
identifying outliers, comparing distributions between different groups, and providing a quick overview of data symmetry or skewness.
```

```python
plt.boxplot(box1,box2)
```
