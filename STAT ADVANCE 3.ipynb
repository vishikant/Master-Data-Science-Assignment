```python
Hypothesis testing is a statistical method used to make inferences about population parameters based on sample data.
It involves formulating a hypothesis about the population parameter, collecting sample data, and using statistical 
tests to assess the strength of evidence against the null hypothesis.

It is used to:
1. Evaluate Claims: Determine whether there is enough evidence to support or reject a claim or hypothesis about a population parameter.
2. Make Decisions: Guide decision-making processes in various fields by providing statistical evidence.
3. Validate Theories: Test theoretical assumptions and hypotheses in scientific research and experimentation.

The importance of hypothesis testing lies in its role in scientific inquiry, decision-making, and drawing 
conclusions based on empirical evidence. It allows researchers and analysts to make informed judgments about 
population parameters, providing a rigorous framework for drawing conclusions from sample data.
```

```python
import scipy.stats as stats

def two_sample_t_test(sample1, sample2):
    # Perform independent two-sample t-test
    t_statistic, p_value = stats.ttest_ind(sample1, sample2)
    return t_statistic, p_value

# Example usage:
sample1 = [83, 87, 90, 78, 82, 85]
sample2 = [75, 80, 88, 72, 79, 84]
t_stat, p_val = two_sample_t_test(sample1, sample2)
print("T-statistic:", t_stat)
print("P-value:", p_val)
```

```python
#Q8. Define p-value and explain its significance in hypothesis testing.
The p-value is the probability of observing the test statistic (or more extreme results) if the null hypothesis 
is true. A low p-value (< 0.05) suggests strong evidence against the null hypothesis, leading to its rejection. 
Conversely, a high p-value suggests weak evidence against the null hypothesis, indicating failure to reject it.
```

```python
import scipy.stats as stats
import math

# Q13: Estimating population mean revenue
sample_mean = 500
sample_std_dev = 50
sample_size = 50
confidence_level = 0.95

# Calculate the critical value (Z) for the confidence level
critical_value = stats.norm.ppf(1 - (1 - confidence_level) / 2)

# Calculate the margin of error
margin_of_error = critical_value * (sample_std_dev / math.sqrt(sample_size))

# Calculate the confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print("95% Confidence Interval for population mean revenue: (${:.2f}, ${:.2f})".format(lower_bound, upper_bound))

# Q14: Hypothesis test for blood pressure decrease
null_hypothesis_mean = 10
sample_mean_bp_decrease = 8
sample_std_dev_bp_decrease = 3
sample_size_bp_decrease = 100
alpha = 0.05

# Calculate the t-statistic
t_statistic = (sample_mean_bp_decrease - null_hypothesis_mean) / (sample_std_dev_bp_decrease / math.sqrt(sample_size_bp_decrease))

# Calculate the critical t-value
critical_t_value = stats.t.ppf(alpha, df=sample_size_bp_decrease - 1)

# Perform the hypothesis test
if t_statistic < critical_t_value:
    print("Reject the null hypothesis. There is sufficient evidence to conclude that the new drug decreases blood pressure by more than 10 mmHg.")
else:
    print("Fail to reject the null hypothesis. There is not enough evidence to conclude that the new drug decreases blood pressure by more than 10 mmHg.")
```

```python
# Q15: Hypothesis test for product weight
null_hypothesis_mean_weight = 5
sample_mean_weight = 4.8
sample_std_dev_weight = 0.5
sample_size_weight = 25
alpha_weight = 0.01

# Calculate the t-statistic
t_statistic_weight = (sample_mean_weight - null_hypothesis_mean_weight) / (sample_std_dev_weight / math.sqrt(sample_size_weight))

# Calculate the critical t-value for a one-tailed test
critical_t_value_weight = stats.t.ppf(alpha_weight, df=sample_size_weight - 1)

# Perform the hypothesis test
if t_statistic_weight < critical_t_value_weight:
    print("Reject the null hypothesis. There is sufficient evidence to conclude that the true mean weight of the products is less than 5 pounds.")
else:
    print("Fail to reject the null hypothesis. There is not enough evidence to conclude that the true mean weight of the products is less than 5 pounds.")

# Q16: Hypothesis test for group means
mean_group1 = 80
std_dev_group1 = 10
n1 = 30

mean_group2 = 75
std_dev_group2 = 8
n2 = 40

alpha_groups = 0.01

# Calculate the pooled standard deviation
pooled_std_dev = math.sqrt(((n1 - 1) * std_dev_group1 ** 2 + (n2 - 1) * std_dev_group2 ** 2) / (n1 + n2 - 2))

# Calculate the t-statistic
t_statistic_groups = (mean_group1 - mean_group2) / (pooled_std_dev * math.sqrt(1/n1 + 1/n2))

# Calculate the critical t-value for a two-tailed test
critical_t_value_groups = stats.t.ppf(1 - alpha_groups / 2, df=n1 + n2 - 2)

# Perform the hypothesis test
if abs(t_statistic_groups) > critical_t_value_groups:
    print("Reject the null hypothesis. There is sufficient evidence to conclude that the population means for the two groups are not equal.")
else:
    print("Fail to reject the null hypothesis. There is not enough evidence to conclude that the population means for the two groups are not equal.")

# Q17: Estimating population mean number of ads watched
sample_mean_ads = 4
sample_std_dev_ads = 1.5
sample_size_ads = 50
confidence_level_ads = 0.99

# Calculate the critical value (Z) for the confidence level
critical_value_ads = stats.norm.ppf(1 - (1 - confidence_level_ads) / 2)

# Calculate the margin of error
margin_of_error_ads = critical_value_ads * (sample_std_dev_ads / math.sqrt(sample_size_ads))

# Calculate the confidence interval
lower_bound_ads = sample_mean_ads - margin_of_error_ads
upper_bound_ads = sample_mean_ads + margin_of_error_ads

print("99% Confidence Interval for population mean number of ads watched: ({:.2f}, {:.2f})".format(lower_bound_ads, upper_bound_ads))
```
